```python

# Buat program untuk minta input jumlah baris dan buat
# rangkaian berikut ini
# *
# **
# ***
# ****
# Dan seterusnya sejumlah baris yang diinputkan
# Gunakan for loop dengan range

a = input("masukkan angka : ")
for x in range(int(a)) :
    for y in range(x+1) :
        print("*", end="")
        y+=1
    print()
    x+=1
```
